```python
def crop_image_by_rate(img_path, _x1, _x2, _y1, _y2):
    img = cv2.imread(img_path)
    img_h = img.shape[0]
    img_w = img.shape[1]
    x1, x2, y1, y2 = round(_x1 * img_w),\
                     round(_x2 * img_w),\
                     round(_y1 * img_h),\
                     round(_y2 * img_h)

    # クロップ用の座標取得
    bbox_w, bbox_h = x2 - x1, y2 - y1
    dst = img[y1:y1+bbox_h, x1:x1+bbox_w]
    return dst
```
